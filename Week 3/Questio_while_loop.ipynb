```python
# robust input loop to avoid ValueError on empty/invalid input
while True:
    s = input("Enter A number: ").strip()
    if s == "":
        print("No input provided. Please enter an integer.")
        continue
    try:
        num = int(s)
        break
    except ValueError:
        print("Invalid input. Please enter a valid integer.")

fact = 1
if num < 0:
    print("Not Defined for negative numbers")
else:
    # compute factorial (0! = 1)
    while num > 0:
        fact *= num
        num -= 1
    print("Factorial is:", fact)

    
```
